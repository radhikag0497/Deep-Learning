```python
Features = dataframe.drop(['Survived','Name'], axis=1).values
# Features.head()

targets = dataframe['Survived'].values
```

```python
Features_train, targets_train = Features[:710], targets[:710]

Features_test, targets_test = Features[710:], targets[710:]
```

```python
model = GaussianNB()
model.fit(Features_train, targets_train)

targets_pred = model.predict(Features_test)

# for x in zip(targets_test, targets_pred):
#     print('Expected: {}, Predicted: {}'.format(x[0], x[1]))

accuracy = model.score(Features_test, targets_test)
print('Accuracy:', accuracy)
```
